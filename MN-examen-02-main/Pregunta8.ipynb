```python
import numpy as np

def resolver_sistema(A, b):

    sol = np.linalg.solve(A, b)

    m = -A[1, 0] / A[1, 1]
    return sol, m

# Ejercicio 1
A1 = np.array([[2, 1], [3, 2]])
b1 = np.array([5, 8])
sol1, m1 = resolver_sistema(A1, b1)

# Ejercicio 2
A2 = np.array([[1, 1], [-2, 1]])
b2 = np.array([6, 0])
sol2, m2 = resolver_sistema(A2, b2)

print(f"Ej 1: x={sol1[0]}, y={sol1[1]}, m={m1}")
print(f"Ej 2: x={sol2[0]}, y={sol2[1]}, m={m2}")
```
