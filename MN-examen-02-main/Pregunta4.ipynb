```python
import numpy as np
from scipy.optimize import bisect, newton

def f(x):
    return x**5 + 5.25*x**4 + 4.125*x**3 - 9.125*x**2 - 14.625*x - 5.625


coeffs = [1, 5.25, 4.125, -9.125, -14.625, -5.625]
approx_roots = np.unique(np.real(np.roots(coeffs)[np.isreal(np.roots(coeffs))]).round(4))


raices = []
for r in approx_roots:
    res = newton(f, r)
    raices.append(round(res, 2))

raices = sorted(list(set(raices)))

for i in range(1, 7):
    val = raices[i-1] if i <= len(raices) else "NaN"
    print(f"x{i} = {val}")
```
